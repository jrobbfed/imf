```python
%matplotlib inline
from imf import *
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
```

#### Sample 50,000 and 1,000 solar mass galaxies to represent Hercules and Willman I UFDs:

```python
ns_mmin, ns_mmax = 3., 8.
N_ns_H = []
for i in range(200):
    Her_masses = sample_imf(50000.)
    N_ns_H.append(np.size(Her_masses[(Her_masses > ns_mmin) & (Her_masses < ns_mmax)]))
```

```python
salpeter = Salp()
salpeter(Her_masses)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].hist(N_ns_H, bins=10, histtype='step', label=r'Hercules (50,000 M$_\odot$)')
axes[0].plot([np.median(N_ns_H), np.median(N_ns_H)], [0, 50])
axes[0].text(np.median(N_ns_H)+1, 40, 'Median '+'= '+ str(np.median(N_ns_H)))
axes[0].set_xlabel('Number of Neutron Stars')
axes[0].set_ylabel('count')
axes[0].legend(fontsize=10)
axes[0].set_title('Number of Neutron Stars in 200 realizations of Salpeter IMF', fontsize=10)

n, bins, patches = axes[1].hist(np.log10(Her_masses), 30, histtype='step', lw=3, log=True)
bin_width = bins[1]-bins[0]
bin_mids = bins[:-1] + bin_width/2.

#Normalize the sampled distribution to a salpeter function, to check the sampling
def salp_fit(mass, norm, **kwargs):
    #integ_form=True indicates that the sampling has been binned logarithmically
    return norm * salpeter(mass, integ_form=True)

popt, pcov = optimize.curve_fit(salp_fit, 10**(bin_mids), n)
print popt, pcov
fit = np.poly1d(z)
#axes[1].plot(np.log10(Her_masses[0::10]), 10.**fit(np.log10(Her_masses[0::10])))
axes[1].plot(np.log10(Her_masses[0::10]), salp_fit(Her_masses[0::10], popt), ':', lw=2, label='Salpeter')
axes[1].legend(fontsize=12)

axes[1].set_ylabel('Count')
axes[1].set_xlabel(r'log(Mass [M$_\odot$])')
fig.set_size_inches(12,6)
plt.savefig('Her_ns.pdf')
```
